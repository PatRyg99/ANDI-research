```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = ".data/datasets/mily-data/Andi_dane/"
y_train = np.argmax(np.load(data_path + "y_train.npy"), axis=1)
X_train = np.load(data_path + "X_train.npy")
```

```python
def plot_samples(X_train, y_train, classes, samples=4):
    fig, ax = plt.subplots(nrows=len(classes), ncols=samples, figsize=(20, 20))

    for i, cls in enumerate(classes):
        class_data = X_train[y_train == cls]

        sampled_indices = np.random.choice(len(class_data), size=samples)
        sampled_class_data = class_data[sampled_indices]

        for j in range(samples):
            ax[i, j].plot(*sampled_class_data[j].T, "-", zorder=1)
            ax[i, j].scatter(*sampled_class_data[j, 0], c="green", s=20, zorder=2)
            ax[i, j].scatter(*sampled_class_data[j, -1], c="red", s=20, zorder=2)
            ax[i, j].set_title(f"Class {cls}")

    plt.tight_layout()
    plt.show()
```

```python
plot_samples(X_train, y_train, classes=[0, 1, 2, 3, 4])
```
